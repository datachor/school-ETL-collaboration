```python
from sqlalchemy import create_engine, inspect
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
import pandas as pd
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/school_db')
conn = engine.connect()
session = Session(engine)
```

```python
# Able to connect to the DB and import a table into Pandas
data = pd.read_sql("SELECT * FROM school_distinctive_programs", conn)
data.head()
```

```python
data_refined3_df = data.loc[data['alp_domain'] == 'STEM', :]
data_refined3_df
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
SLP = Base.classes.school_distinctive_programs
GIOS = Base.classes.general_information_of_schools
```

```python
# Query for All the distinct STEM schools
stem_schools = session.query( SLP.school_name, SLP.alp_domain ).\
    filter( SLP.alp_domain == 'STEM')

# stem_schools.all()
# Creating a list of all the schools
school_list = []
for school in stem_schools:
    school_list.append(school[0])

school_list
```

```python
titles = pd.DataFrame(
    {
        "employee_title": ["Principal", "VP"]
    }
)
titles.to_sql("titles", conn, index=True, index_label="title_id")
```

```python
# Transforming the columns in GIOP for a "Titles" table
admin_data = session.query( GIOS.school_name, GIOS.principal_name, GIOS.first_vp_name, GIOS.second_vp_name, GIOS.third_vp_name, GIOS.fourth_vp_name, GIOS.fifth_vp_name, GIOS.sixth_vp_name )

# Creating a list to create a data frame
admin_list = []

# Commands to look at the data from the query
# admin_data.all()
# admin_data[0]


for admin in admin_data:
    s = admin[0]
    # print(s)
    for x in range(1, 8):
        person = []
        a = admin[x]
        if x == 1:
            t = '0'
        if x > 1:
            t = '1'
        if admin[x] == 'NULL':
            break        
        person.append(admin[0])
        person.append(a)
        person.append(t)
        admin_list.append(person)

admin_table = pd.DataFrame(admin_list, columns=["school_name", 'name', 'title'])
admin_table.to_sql("admins", conn)
# Checking to see that all Six VPs got into the table
# admin_table.loc[admin_table[0] == 'RAFFLES INSTITUTION']
```

```python
# Transforming the columns in GIOP for a "Mother Tongue" table
mt_data = session.query( GIOS.school_name, GIOS.mothertongue1_code, GIOS.mothertongue2_code, GIOS.mothertongue3_code )

mother_tongue_data = []

for mt in mt_data:
    s = mt[0]
    # print(s)
    for x in range(1, 4):
        tongue = []
        a = mt[x]
        if mt[x] == 'na':
            break        
        tongue.append(mt[0])
        tongue.append(a)
        mother_tongue_data.append(tongue)

mt_df = pd.DataFrame(mother_tongue_data, columns=['school_name', 'mothertongue_code'])
mt_df.to_sql("mother_tongue_data", conn)
```

```python
prg_data = session.query( SLP.school_name, SLP.alp_domain, SLP.alp_title, SLP.llp_domain1, SLP.llp_title1, SLP.llp_domain2, SLP.llp_title2 )


for admin in prg_data:
    s = admin[0]
    # print(s)
    for x in range(1, 8):
        person = []
        a = admin[x]
        if x == 1:
            t = 'Principal'
        if x > 1:
            t = 'Vice Principal'
        if admin[x] == 'NULL':
            break        
        person.append(admin[0])
        person.append(a)
        person.append(t)
        admin_list.append(person)
```
